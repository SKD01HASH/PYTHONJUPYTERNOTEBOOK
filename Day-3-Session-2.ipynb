#Break

```
# Read 100 integer inputs from a user.
# Print the sum of inputs until a negative input is found (Excluding the negative number)
# or all 100 inputs are exhausted.

sum = 0

for i in range(100):

  value = int(input())
  if (value < 0):
    # if the number is negative, exit from the loop
    break
  sum += value

print('sum =', sum)
```

```
# Read 100 integer inputs from a user.
# Print the sum of inputs until a negative input is found
#(Including the negative number)
# or all 100 inputs are exhausted.

sum = 0
for i in range(100):

  value = int(input())
  sum += value
  if (value < 0):
    # if the number is negative, exit from the loop
    break
print('sum =', sum)
```

```
# Read 100 integer inputs from a user.
# Print the sum of inputs until a negative input is found
# and add 2 times the negative number.
# or all 100 inputs are exhausted.

sum = 0

for i in range(100):
  value = int(input())
  sum += value

  if (value < 0):
      # if the number is negative, exit from the loop
      sum += value
      break

print('sum =', sum)
```
