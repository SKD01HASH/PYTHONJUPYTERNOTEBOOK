```
# operations on a tuple

tuple1 = 'Akash', 34, 'CSE'
tuple2 = 'Aman', 35, 'ECE'

print(tuple1)
print(tuple1 + tuple2) # concatenation
print((tuple1 + tuple2)[4]) # acessing an element of a tuple
print((tuple1 + tuple2)[1:4]) # slicing
print((tuple1 + tuple2) * 3) # repeating values in a tuple
```

```
tuple1[1:50]
```

```
print(tuple1 + tuple1)
s1 = (5,)
s2 = (5)

print(s1 * 3, s2 * 3)
print(type(s1), type(s2))
```

```
# note that tuples are immutable - once created, 
# their members can not be changed.
# We can change the binding of the variable though
#tuple1[0] = 'Aakash'
print(tuple1)
tuple1 = tuple1 + tuple2
print(tuple1)
```

##UNPACKING TUPLES

```
#Unpacking Tuples: Basic
x, y, z = ("apples", 10, "oranges") 
  
# print details 
print(x) 
print(y) 
print(z) 
```

```
#Unpacking Tuples: Advanced
*x, y, z = (10, "apples", 'for', 10, "oranges") 
  
# print details 
print(x) 
print(y) 
print(z) 
```

```
x, y, *z = (10, "apples", 'for', 10, "oranges") 
  
# print details 
print(x) 
print(y) 
print(z) 
```

```
x, *y, z = (10, "apples", 'for', 10, "oranges") 
  
# print details 
print(x) 
print(y) 
print(z) 
```

```
*x, y, *z = (10, "apples", 'for', 10, "oranges") 
  
# print details 
print(x) 
print(y) 
print(z) 
```

```
*x, y, z = (10, "oranges") 
# print details 
print(x) 
print(y) 
print(z) 
```
