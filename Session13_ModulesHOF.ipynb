#Lambda Functions

```python
def summation(f, n):
    ''' compute f(1) + f(2) + f(3) + ... + f(n)'''
    sum = 0
    for i in range(1,n+1):
        sum += f(i)
    return sum

def identity(x):
    return x

def square(x):
    return x*x

def cube(x):
    return x*x*x

s1 = summation(identity, 10)
s2 = summation(square, 10)
s3 = summation(cube, 10)
print(s1, s2, s3, sep='\n')
```

```python
print(summation(lambda x:x, 10))
print(summation(lambda x:x*x, 10))
print(summation(lambda x:x*x*x, 10))
```

```python
v = (lambda x, y: y*y - 2*x)(3,4)
print('Result of lambda', v)
ii = int(input())
if (ii > 100):
    myfun = lambda x, y: y*y - 2*x  # def myfun(x,y): return y*y - 2*x 
    print('Lambda assigned to a variable', myfun(3,4))

list_of_funs = [myfun, summation, identity, lambda x:x*x, 5, 3.14]
print(list_of_funs)
```
