## Strings : Basics

## Strings: Concat and Repeat

```python
# operations on strings

first_phrase = "hello!.. "
second_phrase = "Hello! Who is this?"
third_phrase = "I won't fly paper airplanes in the class!"

print(first_phrase * 3) # repeats the string three times

fs = first_phrase + second_phrase # concatenates the two strings
print(fs)

print(4 * third_phrase ) # repeats the string four times
```

## Strings: Indexing and Slicing

```python
# indexing

name = 'Python Programming'

print(name[0]) # print the first character in the string

print(name[3]) # print the fourth character in the string
```

```python
'''
What is the result of below program?  a. obj; b. str; c. char
'''
c = name[3]
type(c)
```

```python
print(name[-6]) # get sixth character from the right side
```

```python
print(name[-1]) # print the last character

print(name[-2]) # print the second last character
```

```python
# Slice / Substring

# print first three letters
print(name[0:3])
# alt. way
print(name[:3]) 

# print letters from index 1 to index 4
print(name[1:5])
```

```python
# Can we pass slice of a string to len() method?

print("Length of slice *{}* is {}".format(name[1:15], len(name[1:15])))
```

```python
print(name[:]) # string containing all the characters

# equivalent to 

print(name[0:len(name)])
```

```python
print(name[-4:-1]) # substring from fourth last character to the second last character from end

print(name[-4:]) # substring from fourth last character of string

print(name[-14:14])
```

```python
print(name, len(name))
```

```python
name[1:70] # note that the original string has only 18 characters
```

```python
name[17:1]
```

```python
name[-1:1:-1]
```

## Strings: Some Other Functions

```python
title = 'python program.'

# Convert first character of the string to upper case
title.capitalize()
```

```python
title = 'Python Programming Is Fun.'

# Convert the string to lower case
title.casefold()
```

```python
# count how many times a letter occurs in a string
title.count('n')
```

```python
# get the position of a character in a string
title.find('y')
```

```python
# What happens if we pass a character that is not present in the string?
title.find('C')
```

```python
# What happens if we pass a character that has multiple occurences in the string?
title.find('n')
```

```python
# Rightmost occurence
k = title.rfind('n')
print(k)
print(title[k-2:k+3])
```

```python
# check if the string ends with a specified value
title.endswith('Fun.')
```

```python
name = '1st Python Programming Course !'

name.isalnum() # checks if all characters in a string are alphaneumeric
```

```python
# lets try to get rid of '!'
# we replace '!' with nothing (empty string).
print(name)
print(name.replace('!', ''))
print(name)
```

```python
# lets try again 
# we replace '!' with '.'

name = name.replace('!', '.')
print(name)
```

```python
# check if all characters in the string are:
# uppercase or lowercase
# or if the string follow all rules of a title

name = 'Python Programming Course!'

print(name.isupper())

print(name.islower())

print(name.istitle())
```

```python
# join a sequence of elements in an iterable with a '-'


print('-'.join(['Welcome', 'to', 'this', 'course!']))
```

```python
string = 'var1'

# check if the string is an identifier (?)
string.isidentifier()
```

```python
string = '1var'

# check if the string is an identifier (?)
string.isidentifier()
```
