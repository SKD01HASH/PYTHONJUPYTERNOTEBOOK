## Using exception to "fix" unwanted scenario

```
x = int(input())
```

```
def read_int():
    try:
        x = int(input())
    except Exception as e: 
        print('Bad Input. Caused Exception', e, ' Try Again: ')
        x = read_int()
        return x
    else:
        return x
    
```

```
x = read_int()
print(x)
```

## Order of except clauses: Does it matter?

```
# Does the except ... order matter? [1/4]
try:
    5/0
except ArithmeticError:
    print("Bad Divison")
except ZeroDivisionError:
    print("Divison by Zero")
```

```
# Does the except ... order matter? [2/4]
try:
    5/0
except ZeroDivisionError:
    print("Divison by Zero")
except ArithmeticError:
    print("Bad Divison")
```

```
# Does the except ... order matter? [3/4]
try:
    5/0
except Exception:
    print('Catch All')
except ArithmeticError:
    print("Bad Divison")
except ZeroDivisionError:
    print("Divison by Zero")
```

```
# Does the except ... order matter? [4/4]
import math
try:
    # "hello"[50]
    5/ 0
    math.sqrt(-1)
    # 4 + "5"
except ZeroDivisionError as z:
    print("Divison by Zero:", z)
except ValueError as v:
    print("I am a value error: ", v)
except Exception as e:
    print('Catch All:', e, type(e))
```

## raising our own Exception

```
# Dissecting an Exception: Example from Python Documentation
#      https://docs.python.org/3/tutorial/errors.html
try:
    raise Exception('spam', 'eggs')
except Exception as inst:
    print(type(inst))    # the exception instance
    print(inst.args)     # arguments stored in .args
    print(inst)          # __str__ allows args to be printed directly,
                         # but may be overridden in exception subclasses
    x, y = inst.args     # unpack args
    print('x =', x)
    print('y =', y)
```

```
# Code to read the age of a person
def read_age():
    age = read_int()
    if age <= 0 or age >= 100:
        raise Exception(age)
    else:
        return age
    
```

```
# Business logic
good, bad, n = 0, 0, 0
while (n < 5):
    n += 1
    try:
        age = read_age()
        good += 1
    except Exception as e:
        bad += 1 
    
print('Read {} good and {} bad ages.'.format(good, bad))
```

```
# Neted Exceptions!!!
try:
    5/0
except ZeroDivisionError as z:
    try:
        raise Exception("My exception 123")
    except Exception as e :
        print("All is Well, even though z {} and e {} were raised".format(z,e))
except ArithmeticError:
    print("Bad Divison")
```

```
# no check

def compute_avg(marks):
    '''computes the average of marks'''
    sum = 0.0
    for m in marks:
        sum = sum + m
    return sum/len(marks)
    # what if marks happen to be empty?
```

```
b = compute_avg([7, 7, 8])
print(b)
```

```
# check, without assert

def compute_avg(marks):
    '''computes the average of marks
       marks cannot be empty.'''
    if len(marks) > 0 :
        sum = 0.0
        for m in marks:
            sum = sum + m
        return sum/len(marks)
    # what to do if marks happen to be empty?
```

```
b = compute_avg([7, 7, 8])
print(b)
```

```
a = compute_avg([])
print(a)
```

```
# assert
def compute_avg(marks):
    '''computes the average of marks
        marks cannot be empty.'''
    assert len(marks) > 0, 'Average Undefined'
    sum = 0.0
    for m in marks:
        sum = sum + m
    return sum/len(marks)
```

```
b = compute_avg([7, 7, 8])
print(b)
```

```
a = compute_avg([])
print(a)
```
